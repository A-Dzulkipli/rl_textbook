```python
def extract_regret(num_arms, epsilon, iterations, episodes):
    regret = []
    for _ in range(episodes):
        bandit_test_bed = BanditTestBed(num_arms, EpsilonGreedy(num_arms, epsilon))
        rewards, arm_choices, arms, best_arm, average_reward = bandit_test_bed.run(iterations)
        regret.append([arms[best_arm] - arms[choice] for choice in arm_choices])
    return regret
```

```python
data = extract_regret(10, 0.1, 1000, 1000)
```

```python
average_regret = [np.mean([data[i][j] for i in range(len(data))]) for j in range(len(data[0]))]
```

```python
plt.plot(average_regret)
```
